```python
import sys  # Always needed
print("Python version:", sys.version)
import pandas as pd
print("pandas version:", pd.__version__)
import matplotlib
print("matplotlib version:", matplotlib.__version__)
import numpy as np
print("NumPy version:", np.__version__)
import scipy as sp
print("SciPy version:", sp.__version__)
import IPython
print("IPython version:", IPython.__version__)
import sklearn
print("scikit-learn version:", sklearn.__version__)
import mglearn
import os
```

```python

dataset = pd.read_csv("Files\season-1718_csv.csv", parse_dates=["Date"], usecols=['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR'])
dataset.iloc[:5]

team_names = {'Manchester City': 0, 'Manchester United': 1, 'Tottenham Hotspur': 2, 'Liverpool': 3, 'Chelsea': 4, 'Arsenal': 5, 'Burnley': 6, 'Everton': 7, 'Leicester City': 8, 'Newcastle': 9, 'Crystal Palace': 10, 'Bournemouth': 11, 'West Ham': 12, 'Watford': 13, 'Brighton': 14, 'Huddersfield': 15, 'Southampton': 16, 'Swansea': 17, 'Stoke City': 18, 'West Brom': 19}

feature_names = ['HomeTeam', 'AwayTeam','HomeLastResult', 'AwayLastResult']
target_names = {'H': 2, 'D': 1, 'A': 0}
```

```python

# Creating two new features determining whether each team won their last match.
from collections import defaultdict

last_result = defaultdict(lambda: -1)

dataset["HomeLastResult"] = -1
dataset["AwayLastResult"] = -1

for index, row in dataset.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    row['FTR'] = target_names[row['FTR']]
    row['HomeTeam'] = team_names[home_team]
    row['AwayTeam'] = team_names[away_team] 
    dataset.iloc[index] = row
```

```python
for index, row in dataset.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']  
    row['HomeLastResult'] = last_result[home_team]
    row['AwayLastResult'] = last_result[away_team]  
    dataset.iloc[index] = row
    if row['FTR'] == 2:
        last_result[home_team] = 2
        last_result[away_team] = 0
    elif row['FTR'] == 0:
        last_result[home_team] = 0
        last_result[away_team] = 2
    elif row['FTR'] == 1:
        last_result[home_team] = 1
        last_result[away_team] = 1
    else:
        last_result[home_team] = -1
        last_result[away_team] = -1
        
        
 # needs to be not but we have 3 classes not true/false
  
    #myrow = dataset.loc[[index]]
    #display(myrow)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset[feature_names].values, dataset['FTR'].values, shuffle=False, random_state = 5)
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)

X_train[0:20]
```

```python
y_train[0:20]
y_train = y_train.astype('int')
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=32)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
# Using the team names and both team's last result as features.
print("Test set score:  {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
for index, row in standings.iterrows():
    team = row["Team"]   
    if team == "Hull City":
        row['Team'] = "Newcastle"  
    if team == "Middlesbrough":
        row['Team'] = "Brighton"    
    if team == "Sunderland":
        row['Team'] = "Huddersfield"
    
    standings.iloc[index] = row
standings
```

```python
# Encode our standings dataset in order to make comparisons with our main dataset

for index, row in standings.iterrows():
    currentTeam = row["Team"]
    for key, value in team_names.items():
        if key == currentTeam:
            row['Team'] = value
            print(currentTeam)
            print(row['Team'])
            standings.iloc[index] = row
standings
```

```python
dataset['HomeTeamRanksHigher'] = 0
for index, row in dataset.iterrows():
    home_team = row["HomeTeam"]
    away_team = row["AwayTeam"]
                    
    home_rank = standings[standings["Team"] == home_team]["Rank"].values[0]
    print("Home rank: ", home_rank)
    away_rank = standings[standings["Team"] == away_team]["Rank"].values[0]
    print("Away rank: ", away_rank)
    row['HomeTeamRanksHigher'] = int(home_rank > away_rank)
    dataset.iloc[index] = row
```
