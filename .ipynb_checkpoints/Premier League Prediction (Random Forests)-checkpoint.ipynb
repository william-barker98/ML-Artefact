# Loading the Dataset

```python
import sys  # Always needed
print("Python version:", sys.version)
import pandas as pd
print("pandas version:", pd.__version__)
import matplotlib
print("matplotlib version:", matplotlib.__version__)
import numpy as np
print("NumPy version:", np.__version__)
import scipy as sp
print("SciPy version:", sp.__version__)
import IPython
print("IPython version:", IPython.__version__)
import sklearn
print("scikit-learn version:", sklearn.__version__)
import os
```

```python

dataset = pd.read_csv("Files\season-1718_csv.csv", parse_dates=["Date"])
dataset.iloc[:5]
```

```python


dataset["HomeWin"]= dataset["FTAG"] < dataset["FTHG"]
y_true = dataset['HomeWin'].values
y_true
```

```python
from collections import defaultdict
last_win = defaultdict(int)
dataset["HomeLastWin"] = False
dataset["AwayLastWin"] = False
```

```python
for index, row in dataset.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    row['HomeLastWin'] = last_win[home_team]
    row['AwayLastWin'] = last_win[away_team]
    dataset.iloc[index] = row      
    last_win[home_team] = row['HomeWin']
    last_win[away_team] = row['HomeWin']

    
```

```python
dataset.iloc[20:25]
```

# Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=5)
```

```python
X_previouswins = dataset[['HomeLastWin', 'AwayLastWin']].values
X_previouswins
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X_previouswins, y_true,
    scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
# Win streaks?
dataset["HomeWinStreak"], dataset["AwayWinStreak"] = 0, 0
win_streak = defaultdict(int)

for index, row in dataset.iterrows():
    home_team = row["HomeTeam"]
    away_team = row["AwayTeam"]
    row["HomeWinStreak"] = win_streak[home_team]
    row["AwayWinStreak"] = win_streak[away_team]
    dataset.iloc[index] = row    
    # Set current win
    if row["HomeWin"]:
        win_streak[home_team] += 1
        win_streak[away_team] = 0
    else:
        win_streak[home_team] = 0
        win_streak[away_team] += 1
    
```

```python
clf = DecisionTreeClassifier(random_state=5)
X_winstreak =  dataset[["HomeLastWin", "AwayLastWin", "HomeWinStreak", "AwayWinStreak"]].values
scores = cross_val_score(clf, X_winstreak, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
standings = pd.read_csv("Files\epl20162017.csv")
```

```python
standings
```

```python
dataset['HomeTeamRanksHigher'] = 0
for index, row in dataset.iterrows():
    home_team = row["HomeTeam"]
    away_team = row["AwayTeam"]
    
    if home_team == "Newcastle":
        home_team = "Hull City"
    elif away_team == "Newcastle":
        away_team = "Hull City"
        
    if home_team == "Brighton":
        home_team = "Middlesbrough"
    elif away_team == "Brighton":
        away_team = "Middlesbrough"
        
    if home_team == "Huddersfield":
        home_team = "Sunderland"
    elif away_team == "Huddersfield":
        away_team = "Sunderland"
                    
                                  
                    
    home_rank = standings[standings["Team"] ==
        home_team]["Rank"].values[0]
    print("Home rank: ", home_rank)
    away_rank = standings[standings["Team"] ==
        away_team]["Rank"].values[0]
    print("Away rank: ", away_rank)
    row['HomeTeamRanksHigher'] = int(home_rank > away_rank)
    dataset.iloc[index] = row
```

```python
X_homehigher = dataset[["HomeLastWin", "AwayLastWin",
"HomeTeamRanksHigher"]].values
```

```python
clf = DecisionTreeClassifier(random_state=5)
scores = cross_val_score(clf, X_homehigher, y_true,
scoring='accuracy')
# Using the last result from both teams
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
last_match_winner = defaultdict(int)
dataset["HomeTeamWonLast"] = 0
```

```python
for index, row in dataset.iterrows():
    home_team = row["HomeTeam"]
    visitor_team = row["AwayTeam"]
    teams = tuple(sorted([home_team, visitor_team]))
    if last_match_winner[teams] == row["HomeTeam"]:
        row["HomeTeamWonLast"] = 1
    else:
        row["HomeTeamWonLast"] = 0
    dataset.iloc[index] = row
```

```python
winner = row["HomeTeam"] if row ["HomeWin"] else row["AwayTeam"]
last_match_winner[teams] = winner
```

```python
X_lastwinner = dataset[["HomeTeamRanksHigher", "HomeTeamWonLast"]].values
clf = DecisionTreeClassifier(random_state=5)
scores = cross_val_score(clf, X_lastwinner, y_true,
scoring='accuracy')
# Using which team ranks higher from last season
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

# Preprocessing

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encoding = LabelEncoder()
display(encoding.fit(dataset["HomeTeam"].values))

# We extract all of the labels for the home teams and visitor teams, and then join them
home_teams = encoding.transform(dataset["HomeTeam"].values)
away_teams = encoding.transform(dataset["AwayTeam"].values)
X_teams = np.vstack([home_teams, away_teams]).T

onehot = OneHotEncoder()
X_teams = onehot.fit_transform(X_teams).todense()

clf = DecisionTreeClassifier(random_state=5)
scores = cross_val_score(clf, X_teams, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))

```

# Random Forests

```python
# This may result in less accuracy, but is able to generalize better than using DecisionTrees

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=5)
scores = cross_val_score(clf, X_teams, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))

```

```python
X_all = np.hstack([X_homehigher, X_teams])
clf = RandomForestClassifier(random_state=5)
scores = cross_val_score(clf, X_all, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```
